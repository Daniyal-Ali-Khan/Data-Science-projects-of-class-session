```python
# Step 1: Load the Data
import pandas as pd

# Read the CSV file
df = pd.read_csv('exams.csv')

# Step 2: Data Summary
# Display the first few rows of the DataFrame
print("First few rows of the dataset:")
df.head()
```

```python
# Display the basic information about the dataset
print("\nBasic information about the dataset:")
df.info()
```

```python
# Check for missing values
print("\nMissing values:")
df.isnull().sum()
```

```python
# Describe numerical variables
print("\nSummary statistics for numerical variables:")
df.describe()
```

```python
# Visualize the distribution of numerical variables
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
sns.histplot(df['math score'], kde=True)
plt.title('Distribution of Math Scores')
```

```python
plt.subplot(1, 3, 2)
sns.histplot(df['reading score'], kde=True)
plt.title('Distribution of Reading Scores')
```

```python
plt.subplot(1, 3, 3)
sns.histplot(df['writing score'], kde=True)
plt.title('Distribution of Writing Scores')
```

```python
plt.subplot(1, 3, 1)
sns.scatterplot(x='math score', y='reading score', data=df)
plt.title('Math Score vs Reading Score')
```

```python
plt.subplot(1, 3, 2)
sns.scatterplot(x='math score', y='writing score', data=df)
plt.title('Math Score vs Writing Score')
```

```python
plt.subplot(1, 3, 3)
sns.scatterplot(x='reading score', y='writing score', data=df)
plt.title('Reading Score vs Writing Score')
```

```python
# Step 6: Visualization
# Relationship between categorical variables and numerical variables
plt.figure(figsize=(16, 8))

plt.subplot(2, 3, 1)
sns.boxplot(x='gender', y='math score', data=df)
plt.title('Gender vs Math Score')

plt.subplot(2, 3, 2)
sns.boxplot(x='race/ethnicity', y='math score', data=df)
plt.title('Race/Ethnicity vs Math Score')

plt.subplot(2, 3, 3)
sns.boxplot(x='parental level of education', y='math score', data=df)
plt.xticks(rotation=45)
plt.title('Parental Level of Education vs Math Score')

plt.subplot(2, 3, 4)
sns.boxplot(x='gender', y='reading score', data=df)
plt.title('Gender vs Reading Score')

plt.subplot(2, 3, 5)
sns.boxplot(x='race/ethnicity', y='reading score', data=df)
plt.title('Race/Ethnicity vs Reading Score')

plt.subplot(2, 3, 6)
sns.boxplot(x='parental level of education', y='reading score', data=df)
plt.xticks(rotation=45)
plt.title('Parental Level of Education vs Reading Score')

plt.tight_layout()
plt.show()
```
