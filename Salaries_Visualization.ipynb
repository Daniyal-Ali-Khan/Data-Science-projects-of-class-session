```python
# Load the dataset
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('salaries.csv')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
# Plot count of 'salary_more_then_100k'
plt.figure(figsize=(8, 6))
sns.countplot(x='salary_more_then_100k', data=data)
plt.title('Count of Salary More than $100k')
plt.xlabel('Salary More than $100k')
plt.ylabel('Count')
plt.show()
```

```python
# Plot count of 'salary_more_then_100k' by 'company'
plt.figure(figsize=(10, 6))
sns.countplot(x='company', hue='salary_more_then_100k', data=data)
plt.title('Count of Salary More than $100k by Company')
plt.xlabel('Company')
plt.ylabel('Count')
plt.legend(title='Salary More than $100k', loc='upper right')
plt.show()
```

```python
# Plot count of 'salary_more_then_100k' by 'job'
plt.figure(figsize=(10, 6))
sns.countplot(x='job', hue='salary_more_then_100k', data=data)
plt.title('Count of Salary More than $100k by Job')
plt.xlabel('Job')
plt.ylabel('Count')
plt.legend(title='Salary More than $100k', loc='upper right')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot count of 'salary_more_then_100k' by 'degree'
plt.figure(figsize=(8, 6))
sns.countplot(x='degree', hue='salary_more_then_100k', data=data)
plt.title('Count of Salary More than $100k by Degree')
plt.xlabel('Degree')
plt.ylabel('Count')
plt.legend(title='Salary More than $100k', loc='upper right')
plt.show()
```
